```python
import pandas as pd

kk = pd.read_csv("KK_Vaccines_replication_data.csv")
print(kk)
print("了解是否有遺失值")
print(kk.isnull().sum())
print("了解欄位資料型態")
print(kk.dtypes)
```

```python
print("欄位的內容必須轉換為數字才可以進行各種學習")
print("VAX3欄位是最後的結果，不做OneHot動作")
list1 = list(kk.columns)
print(list1)
list1.pop()
print(list1)
print(kk.shape)
kk2 = kk[list1]
print(kk2.shape)
kk3 = pd.get_dummies(kk2)
print(kk3.shape)
print("最後結果加入")
vax3_mapping = {"Yes": 1, "No": 2}
kk3["ans"] = kk["VAX3"].map(vax3_mapping)
print(kk3.shape)
kk2.to_html("kk2.html")
kk3.to_html("kk3.html")
```
