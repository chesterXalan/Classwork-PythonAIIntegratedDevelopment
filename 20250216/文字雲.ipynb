```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

print("wordcloud讀取jieba的分詞結果，中文資料會有差異")
from collections import Counter

import jieba

text = open("data/mayday1.txt", encoding="utf-8").read()
print("於文字雲呈現之前，了解詞彙")
with open("data/stopword1.txt", encoding="utf-8") as f:
    stops = f.read().split("\n")
terms = []  # 儲存詞彙w
str1 = ""  # 產生一個字串
for t in jieba.cut(text):
    # 如果是半形空格、全形空格、換行符號 那就轉換
    if t == " ":
        t = t.replace(" ", "自訂詞彙1")
        # print('半形空格')
    if t == "\u3000":
        t = t.replace("\u3000", "自訂詞彙2")
        # print('全形空格')
    if t == "\n":
        t = t.replace("\n", "自訂詞彙3")
        # print('換行')
    if t not in stops:
        terms.append(t)
        str1 += "\n" + str(t)
print(str1)  # 預定整個字串分析找出關鍵字，所以每一個詞彙都會做換行
dict1 = Counter(terms)
for x in dict1:
    print(x, dict1[x])
print("str1內容:\n")
print(str1)
print("dict1內容:\n")
print(dict1)
```

```python
stopwords = {}.fromkeys(["沒有", "一個", "什麼", "那個", "oh", "yeah"])
wc = WordCloud(
    font_path="NotoSerifCJKtc-Medium",
    background_color="white",
    max_words=2000,
    stopwords=stopwords,
)
wc.generate(str1)
plt.imshow(wc)
plt.axis("off")
plt.show()
wc.to_file("wordcloud5.jpg")
```
