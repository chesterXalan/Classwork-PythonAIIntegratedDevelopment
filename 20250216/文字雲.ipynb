```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

print("wordcloudjieba")
from collections import Counter

import jieba

text = open("data/mayday1.txt", encoding="utf-8").read()
print("")
with open("data/stopword1.txt", encoding="utf-8") as f:
    stops = f.read().split("\n")
terms = []  # w
str1 = ""  # 
for t in jieba.cut(text):
    #  
    if t == " ":
        t = t.replace(" ", "1")
        # print('')
    if t == "\u3000":
        t = t.replace("\u3000", "2")
        # print('')
    if t == "\n":
        t = t.replace("\n", "3")
        # print('')
    if t not in stops:
        terms.append(t)
        str1 += "\n" + str(t)
print(str1)  # 
dict1 = Counter(terms)
for x in dict1:
    print(x, dict1[x])
print("str1:\n")
print(str1)
print("dict1:\n")
print(dict1)
```
