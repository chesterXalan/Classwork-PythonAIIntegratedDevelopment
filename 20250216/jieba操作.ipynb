```python
# /content/drive/MyDrive/Colab/txt/data.txt
print("jieba載入檔案後分詞，想要知道一個數字")
import jieba

list1 = []
with open("data/data.txt", encoding="UTF-8") as file:
    text = file.read()
    # print('text:',text)
    segments = jieba.cut(text)
    for seg in segments:
        # print(seg)
        list1.append(seg)
print("總共有多少個詞彙:", len(list1))
import collections

counter1 = collections.Counter(list1)
# print(counter1.__class__)
# print(counter1.keys())
# print(counter1.values())
print("找尋前10個詞彙")
for key, values in counter1.most_common(10):
    print(key, ":", values)
print("全部詞彙")
for key, values in counter1.items():
    print(key, ":", values)
```

```python
print("刪除已經定義好的語助詞，這些詞彙你也可以修改")
# /content/drive/MyDrive/Colab/txt/data.txt
import jieba

stopWords = []  # 語助詞清單
segments = []  # 分詞結果
remainderWords = []  # 去除語助詞後的清單
with open("data/stopWords202405_jieba.txt", encoding="UTF-8") as file:
    for data in file:
        data = data.strip()  # 去除空白
        stopWords.append(data)

with open("data/data.txt", encoding="UTF-8") as file:
    text = file.read()
    segments = jieba.cut(text)

remainderWords = list(filter(lambda a: a not in stopWords and a != "\n", segments))
print("總共有多少個詞彙:", len(remainderWords))
import collections

counter1 = collections.Counter(remainderWords)
print("找尋前10個詞彙")
for key, values in counter1.most_common(10):
    print(key, ":", values)
print("全部詞彙")
for key, values in counter1.items():
    print(key, ":", values)
```

```python
print("刪除已經定義好的語助詞，這些詞彙你也可以修改")
import jieba

stopWords = []  # 語助詞清單
segments = []  # 分詞結果
remainderWords = []  # 去除語助詞後的清單
with open("data/stopWords202405_jieba.txt", encoding="UTF-8") as file:
    for data in file:
        data = data.strip()  # 去除空白
        stopWords.append(data)

print("自定義詞彙")
jieba.load_userdict("data/add.txt")

with open("data/data.txt", encoding="UTF-8") as file:
    text = file.read()
    segments = jieba.cut(text)

remainderWords = list(filter(lambda a: a not in stopWords and a != "\n", segments))
print("總共有多少個詞彙:", len(remainderWords))
import collections

counter1 = collections.Counter(remainderWords)
print("找尋前10個詞彙")
for key, values in counter1.most_common(10):
    print(key, ":", values)
print("全部詞彙")
for key, values in counter1.items():
    print(key, ":", values)
```
