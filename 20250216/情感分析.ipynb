```python
from snownlp import SnowNLP

text = "我今天很快樂。我今天很憤怒。"
s = SnowNLP(text)
for sentence in s.sentences:
    print(sentence)
s1 = SnowNLP(s.sentences[0])  # 第一段文字，數值較高代表偏向於正面
print(s1.sentiments)
s1 = SnowNLP(s.sentences[1])  # 第二段文字，數值較高代表偏向於正面
print(s1.sentiments)
```

```python
from snownlp import SnowNLP

s = SnowNLP("這個東西真的很贊")
print(s.words)
print(s.sentiments)

text = """
自然語言處理是計算機科學領域與人工智能領域中的一個重要方向。
它研究能實現人與計算機之間用自然語言進行有效通信的各種理論和方法。
自然語言處理是一門融語言學、計算機科學、數學於一體的科學。
因此，這一領域的研究將涉及自然語言，即人們日常使用的語言，
所以它與語言學的研究有著密切的聯繫，但又有重要的區別。
自然語言處理並不是一般地研究自然語言，
而在於研製能有效地實現自然語言通信的計算機系統，
特別是其中的軟件系統。因而它是計算機科學的一部分。
"""
s = SnowNLP(text)
print(s.sentiments)
print(s.keywords(3))
print(s.summary(3))
```

```python
print("整個就職演說稿進行情感分析")
with open("data/data.txt", encoding="UTF-8") as file:
    text = file.read()
    s = SnowNLP(text)
    print(s.sentiments)
    print(s.keywords(3))
    print(s.summary(3))
```

```python
print("就職演說稿每一段進行情感分析")
with open("data/data.txt", encoding="UTF-8") as file:
    for line1 in file:
        # print(line1)
        s = SnowNLP(line1)
        print(s.sentiments)
        print(s.keywords(3))
        print(s.summary(3))
        print("-" * 100)
```

```python
print("整個就職演說稿進行情感分析:找出最好的一段話")
sentiments1 = 0
keywords1, summary1 = [], []
str1 = ""
with open("data/data.txt", encoding="UTF-8") as file:
    for line1 in file:
        # print(line1)
        s = SnowNLP(line1)
        if s.sentiments > sentiments1:
            sentiments1 = s.sentiments
            keywords1 = s.keywords(3)
            summary1 = s.summary(3)
            str1 = line1
print("最好的分數是:", sentiments1)
print("最好的關鍵字是:", keywords1)
print("最好的詞句是:", summary1)
print("最好的段落是:", str1)
```
